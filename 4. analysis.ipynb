```python
def room(x):
    if x == 'Entire home/apt':
        return 4
    elif x == 'Private room':
        return 3
    elif x  == 'Hotel room':
        return 2
    elif x == 'Shared room':
        return 1
    else:
        return x

path = '/home/xtzie/Desktop/FNLFORM/data/trend/'

periodsummary = []
review_floor = 9
price_ceiling = 3000
screwyewbnb = []
fnlform = []

for file in sorted(os.listdir(path))[0:2]:
    filename = path + file
    if file[1:7] == 'listin':
        df = df[(df.price != 0) & (df.number_of_reviews > review_floor) & (df.price < price_ceiling)]
        test = df.groupby('neighbourhood').mean().sort_values('reviews_per_month', ascending = False).reset_index()
        index_neighbourhood = list(test['neighbourhood'][0:30])  
            
for file in sorted(os.listdir(path)):
    filename = path + file
    if file[1:7] == 'listin':
        df = pd.read_csv(filename)
        df = df[(df.price != 0) & (df.number_of_reviews > review_floor) & (df.price < price_ceiling)]
        df['room_type'] = df['room_type'].map(lambda x: room(x))
        entire_homeorapt = len(df[df.room_type == 4])
        private_room = len(df[df.room_type == 3])
        hotel_room = len(df[df.room_type == 2])
        shared_room = len(df[df.room_type == 1])
        mean = df.price.mean()
        median = df.price.median()
        total_listings = len(df)
        summary = [entire_homeorapt, private_room, hotel_room, shared_room]
        summary = [round(x/total_listings,2) for x in summary] + [total_listings] + [mean] + [median]
        periodsummary += [summary]
        
        entire = round(df[df.room_type == 4]['price'].mean(), 2)
        pvt = round(df[df.room_type == 3]['price'].mean(), 2)
        host = round(df.calculated_host_listings_count.mean(), 2)
        screwyewbnb += [[entire, pvt, host]]
        
        test = df.groupby('neighbourhood').mean().sort_values('reviews_per_month', ascending = False).reset_index()
        test['bookings'] = round(test['reviews_per_month'] * 3, 2)
        test['occupancy'] = round(test['reviews_per_month'] * 3 * 3 /30, 2)
        test['occupancy_adjusted'] = test.occupancy.apply(lambda x: 0.8 if x>.8 else x)
        test['yield'] = test.occupancy_adjusted * test.price * 30
        snapshot = []
        for neigh in index_neighbourhood:
            bookz = round( list(test[test.neighbourhood == neigh]['bookings'])[0], 2)
            ocz_adj = round( list(test[test.neighbourhood == neigh]['occupancy_adjusted'])[0], 2)
            yieldz = round( list(test[test.neighbourhood == neigh]['yield'])[0], 2)
            neighz = [bookz, ocz_adj, yieldz]
            snapshot += [neighz]
        fnlform += [snapshot]
```
